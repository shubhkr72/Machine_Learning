## Matplotlib example workflow

```python
#  0. import to jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1.  Prepare Data
x=[1,2,3,4,5,6]
y=[34,56,32,87,87,5]
# 2. set plot by object-oriented
# figsize x and y
fig,ax=plt.subplots(figsize=(8,5))
# figsize=(width,height)
# 3. Plot data
ax.plot(x,y)
# 4.  customize plot
ax.set(title="Simple Plot",
      xlabel="X-axis",
      ylabel="y-axis")
# 5.  save and show
fig.savefig("sample_plot.png")
```

 # Which one should you use
* When plotting something quickly okay to use pyplot method
* when plotting something more advanced, use OO method

```python
over_50=heart[heart["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",x="age",y="chol")
```

```python
# pyplot method
over_50.plot(kind="scatter",x="age",y="chol",c="target")
```

```python
#OO method mix with pyplot method
fig,ax=plt.subplots(figsize=(10,8))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))
# plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])
# customize the plot
ax.set(title="Heart Disease and cholestrol Levels",xlabel="Age",ylabel="Cholestrol")
# add legend 
ax.legend(*scatter.legend_elements(),title="target");
# add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
 over_50.head()
```

```python
#subplots of chol age thalach
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
#cunstomization
ax0.set(title="Heart dieseas chol level plot",ylabel="chol",)
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="heart Dieseas and Thalach levels",xlabel="Age",ylabel="Thalach")
# add title to figure
fig.suptitle("Heart dieseas Analysis",fontsize=24,fontweight="bold");
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
#cunstomization
ax0.set(title="Heart dieseas chol level plot",ylabel="chol",)
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")
ax1.set(title="heart Dieseas and Thalach levels",xlabel="Age",ylabel="Thalach")
# add title to figure
fig.suptitle("Heart dieseas Analysis",fontsize=24,fontweight="bold");
```

```python
#customizing x and y limits
plt.style.use('seaborn-v0_8-whitegrid')
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#add data to ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
#cunstomization
ax0.set(title="Heart dieseas chol level plot",ylabel="chol")
ax0.set_xlim([50,80])
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="summer")
ax1.set(title="heart Dieseas and Thalach levels",xlabel="Age",ylabel="Thalach")
# add title to figure
fig.suptitle("Heart dieseas Analysis",fontsize=24,fontweight="bold");
```
